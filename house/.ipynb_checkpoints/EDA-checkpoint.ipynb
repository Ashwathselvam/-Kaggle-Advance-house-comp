```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('train.csv')
```

```python
df.head()
```

```python
feature_na = [feature for feature in df.columns if df[feature].isnull().sum()>1]
```

```python
pd.Series(np.array(list(map(lambda x:df[x].isnull().sum(),feature_na))),index=feature_na)
```

```python
for feature in feature_na:
    data=df.copy()
    data[feature] = np.where(data[feature].isnull(),1,0)
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
    
```

```python
from pandas.api.types import is_numeric_dtype
num_fea  = [feature for feature in df.columns if is_numeric_dtype(df[feature])]
num_fea
```

```python

year_feature = [feature for feature in num_fea if 'Yr'in feature or 'Year'in feature]
year_feature
df[year_feature]
```

```python
for fea in year_feature:
    print(fea ,df[fea].unique())
```

```python
df.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year sold')
plt.ylabel('Median house price')
plt.title('House price vs Year sold')
```

```python
for fea in year_feature:
    if fea !='YrSold':
        data=df.copy()
        data[feature]= data['YrSold']-data[fea]
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
        
        
```

```python
con_fea  = [feature for feature in df.columns if df[feature].dtype=='float64']
con_fea
```

```python
for fea in con_fea:
    data = df.copy()
    data[fea].hist(bins=25)
    plt.xlabel(fea)
    plt.ylabel("Count")
    plt.title(fea)
    plt.show()
```

```python
dis_fea = [fea for fea in df.columns if fea not in con_fea]
dis_fea
```

```python
for feature in dis_fea:
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

# LOG  FEATURE

```python
data = df.copy()
for fea in con_fea:
    if 0 in data[fea].unique():
        pass
    else:
        data[fea] = np.log(data[fea])
        data['SalePrice'] = np.log(data['SalePrice'])
        plt.scatter(data[fea],data['SalePrice'])
        plt.xlabel(fea)
        plt.ylabel('SalePrice')
        plt.title(fea)
        plt.show()
    
```
